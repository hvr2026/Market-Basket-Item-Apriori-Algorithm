```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from apyori import apriori
```

```python
# Load data from an Excel file (assuming it's named 'dataset.xlsx')
try:
    data = pd.read_csv('/content/TRAIN-ARULES.csv')
    print("Dataset loaded successfully!")
    # Group products by order_id to form transactions
    transactions = data.groupby('order_id')['product_name'].apply(list).tolist()
    print("Sample transactions:", transactions[:5])
except Exception as e:
    print("An error occurred while loading the dataset:", e)
    raise
```

```python
# Apply the Apriori algorithm
try:
    association_rules = apriori(transactions, min_support=0.0045, min_confidence=0.2, min_lift=1, min_length=2)
    results = list(association_rules)
    print("Number of association rules found:", len(results))
    for rule in results:
        # Extracting the rule antecedents, consequents, support, confidence, and lift
        antecedents = tuple(rule[2][0][0])
        consequents = tuple(rule[2][0][1])
        support = rule[1]
        confidence = rule[2][0][2]
        lift = rule[2][0][3]
        print(f"Rule: {antecedents} -> {consequents}, Support: {support}, Confidence: {confidence}, Lift: {lift}")
except Exception as e:
    print("An error occurred while processing rules with Apriori:", e)
```

```python
# Load and prepare the data from testarules.csv
try:
    test_data = pd.read_csv('/content/testarules.csv')
    print("testarules.csv loaded successfully!")
    print("Columns in testarules.csv:", test_data.columns)  # To verify column names
    # Assuming each row is a transaction and each column is an item if it's non-null
    transactions_test = test_data.apply(lambda row: [item for item in row if pd.notna(item)], axis=1).tolist()
    print("Sample transactions from testarules.csv:", transactions_test[:5])
except Exception as e:
    print("An error occurred while loading or preparing testarules.csv:", e)
```

```python
# Apply the Apriori algorithm to testarules.csv
try:
    if transactions_test:  # Ensure there are transactions to process
        association_rules_test = apriori(transactions_test, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
        results_test = list(association_rules_test)
        print("Number of rules from testarules.csv:", len(results_test))
        for rule in results_test:
            antecedents = tuple(rule[0])
            support = rule[1]
            for detail in rule[2]:
                confidence = detail[2]
                lift = detail[3]
                print(f"Rule: {antecedents} -> {tuple(detail[1])}, Support: {support}, Confidence: {confidence}, Lift: {lift}")
    else:
        print("No transactions available for processing in testarules.csv.")
except Exception as e:
    print("An error occurred while processing testarules.csv with Apriori:", e)
```

```python
try:
    if transactions_test:  # Ensure there are transactions to process
        # Try with very low min_support and min_confidence
        association_rules_test = apriori(transactions_test, min_support=0.0001, min_confidence=0.05, min_lift=1.0, min_length=2)
        results_test = list(association_rules_test)
        print("Adjusted number of rules from testarules.csv:", len(results_test))
        if results_test:
            for rule in results_test:
                antecedents = tuple(rule[0])
                support = rule[1]
                for detail in rule[2]:
                    confidence = detail[2]
                    lift = detail[3]
                    print(f"Rule: {antecedents} -> {tuple(detail[1])}, Support: {support}, Confidence: {confidence}, Lift: {lift}")
        else:
            print("No rules were generated even with very low parameters.")
    else:
        print("No transactions available for processing in testarules.csv.")
except Exception as e:
    print("An error occurred while processing testarules.csv with Apriori:", e)
```
